```python
train=train.drop(columns=["id","CustomerId","Surname"])
test=test.drop(columns=["id","CustomerId","Surname"])
```

```python
train=pd.get_dummies(train,drop_first=True)
test=pd.get_dummies(test,drop_first=True)
```

```python
train.head()
```

```python
X = train.drop("Exited", axis=1)
y = train["Exited"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_valid_scaled=scaler.transform(X_valid)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=1000,random_state=10)
lr.fit(X_train_scaled,y_train)
```

```python
y_pred=lr.predict(X_valid_scaled)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy:",accuracy_score(y_valid,y_pred))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_valid, y_pred))
print(classification_report(y_valid, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train_scaled,y_train)
```

```python
knn_pred=knn.predict(X_valid_scaled)
from sklearn.metrics import accuracy_score
print("Accuracy:",accuracy_score(y_valid,knn_pred))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_valid, knn_pred))
print(classification_report(y_valid, knn_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(random_state=10)
dt.fit(X_train,y_train)
```

```python
dt_pred=dt.predict(X_valid)
from sklearn.metrics import accuracy_score
print("Accuracy:",accuracy_score(y_valid,dt_pred))
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y_valid,dt_pred))
print(classification_report(y_valid,dt_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(random_state=10)
rf.fit(X_train,y_train)
```

```python
rf_pred=rf.predict(X_valid)
from sklearn.metrics import accuracy_score
print("Accuracy:",accuracy_score(y_valid,rf_pred))
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y_valid,rf_pred))
print(classification_report(y_valid,rf_pred))
```

```python
from sklearn.model_selection import GridSearchCV
rf = RandomForestClassifier(random_state=10)
```

```python
param_grid = {
    "n_estimators": [100, 200],
    "max_depth": [10, 20, None],
    "min_samples_split": [2, 5]
}
```

```python
grid = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    scoring="accuracy",
    n_jobs=-1
)
```

```python
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
