```python
null_cols=train.columns[train.isnull().sum()>0]
train[null_cols].dtypes
```

```python
test_ids = test["Id"]
```

```python
null_cols=['Alley', 'MasVnrType', 'BsmtQual', 'BsmtCond', 'BsmtExposure',
    'BsmtFinType1', 'BsmtFinType2', 'FireplaceQu',
    'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',
    'Fence', 'MiscFeature']
for data in[train,test]:
 for col in null_cols:
    train[col]=train[col].fillna('None')
```

```python
train['LotFrontage']=train['LotFrontage'].fillna(train['LotFrontage'].median())
```

```python
train["GarageYrBlt"] = train["GarageYrBlt"].fillna(train["GarageYrBlt"].median())
```

```python
train["MasVnrArea"] = train["MasVnrArea"].fillna( train["MasVnrArea"].median())
```

```python
train['Electrical'] = train['Electrical'].fillna(train['Electrical'].mode()[0])
```

```python
train["TotalBathrooms"] = (
    train["FullBath"] +
    0.5*train["HalfBath"] +
    train["BsmtFullBath"] +
    0.5*train["BsmtHalfBath"]
)
```

```python
train["HouseAge"]=train["YrSold"] - train["YearBuilt"]
```

```python
train = train.drop('PoolQC', axis=1, errors='ignore')
test = test.drop('PoolQC', axis=1, errors='ignore')
```

```python
y=train["SalePrice"]
X=train.drop("SalePrice",axis=1)
```

```python
X = pd.get_dummies(X, drop_first=True)
test = pd.get_dummies(test, drop_first=True)

X, test = X.align(test, join='left', axis=1, fill_value=0)

print(X.isnull().sum().sum())
print(test.isnull().sum().sum())
```

```python
print(test.isnull().sum()[test.isnull().sum() > 0])
```

```python
for col in test.columns:
    if test[col].dtype == 'object':
        test[col] = test[col].fillna(test[col].mode()[0])
    else:
        test[col] = test[col].fillna(test[col].median())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
lr_pred=lr.predict(X_valid)
print("Linear Regression")
print("MAE:", mean_absolute_error(y_valid, lr_pred))
print("R2:", r2_score(y_valid, lr_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100, random_state=10)
rf.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
rf_pred=rf.predict(X_valid)
print("RandomForest")
print("MAE:",mean_absolute_error(y_valid,rf_pred))
print("R2:",r2_score(y_valid,rf_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor(random_state=10)
dt.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
dt_pred=dt.predict(X_valid)
print("DecisionTree")
print("RME:",mean_absolute_error(y_valid,dt_pred))
print("R2:",r2_score(y_valid,dt_pred))
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_valid_scaled=scaler.transform(X_valid)
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor()
knn.fit(X_train_scaled,y_train)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score
```

```python
knn_pred=knn.predict(X_valid_scaled)
print("KNeighborsRegressor")
print("MAE:",mean_absolute_error(y_valid,knn_pred))
print("R2:",r2_score(y_valid,knn_pred))   
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
rf = RandomForestRegressor(random_state=10)
```

```python
param_grid = {
    "n_estimators": [100, 200],
    "max_depth": [10, 20, None],
    "min_samples_split": [2, 5]
}
```

```python
grid = GridSearchCV(
    rf,
    param_grid,
    cv=5,
    scoring="r2",
    n_jobs=-1
)
```

```python
grid.fit(X_train,y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
best_rf = grid.best_estimator_
rf_tuned_pred = best_rf.predict(X_valid)
from sklearn.metrics import mean_absolute_error, r2_score
print("Tuned Random Forest")
print("MAE:", mean_absolute_error(y_valid, rf_tuned_pred))
print("R2:", r2_score(y_valid, rf_tuned_pred))
```

```python
best_rf.fit(X,y)
test_pred=best_rf.predict(test)
```

```python
submission = pd.DataFrame({
    "Id": test_ids,
    "SalePrice": test_pred
})
submission.to_csv("submission.csv", index=False)
submission.head()
```
